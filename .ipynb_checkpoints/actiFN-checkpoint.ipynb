```python
import numpy as np
import matplotlib.pyplot as plt 
from timeit import default_timer as timer
from datetime import timedelta
import math
```

```python
def relu(x):
    return np.maximum(0, x)
def sigmoid(x):
    sig = 1 / (1 + math.exp(-x))
    return sig
def tanh(x):
    return (np.exp(x)- np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
alist  = np.arange(-2.0, 2.0, 0.01)
timing = np.zeros((400))

for i in range(400):
    realcal = (i / 100) - 2.00
#    print(real_cal) 
    start = timer()
    a = sigmoid(realcal)
    end = timer()
    timing[i] = end - start

plt.plot(alist, timing, '.')
plt.xlabel('input')
plt.ylabel('Timing')
plt.grid()
```

```python
point = 80
bb = np.arange(-2.0, 2.0, 0.05)
key = np.zeros((point))
time = 0
for i in bb:
    start = timer()
    sigmoid(i)
    end = timer()
    key[time] = end - start
    time += 1

plt.plot(bb[3:], key[3:] , '.')
plt.xlabel('input')
plt.ylabel('Timing')
plt.ylim()
plt.grid()
```

```python
point = 200
bb = np.arange(-1.0, 1.0, 0.01)
key = np.zeros((point))
time = 0
for i in bb:
    start = timer()
    tanh(i)
    end = timer()
    key[time] = end - start
    time += 1


plt.plot(bb, key, '.')
plt.xlabel('input')
plt.ylabel('Timing')
plt.grid()
```
